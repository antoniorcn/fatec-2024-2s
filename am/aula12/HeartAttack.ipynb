```python
import tensorflow as tf
```

```python
import pandas as pd
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Input, Dense
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
from sklearn.preprocessing import normalize
```

```python
df = pd.read_csv("./sample_data/heart.csv")
df.head()
```

```python
x_input = df[['age', 'sex', 'cp', 'trtbps', 'chol','fbs']]
x_input
```

```python
y_output = df[['output']]
y_output
```

```python
x_normalized = normalize(x_input)
x_normalized
```

```python
modelo = Sequential()
modelo.add( Input ( (6,) ))
modelo.add( Dense(39, activation="relu") )
modelo.add( Dense(6, activation="relu") )
modelo.add( Dense(1, activation="sigmoid") )
modelo.summary()
```

```python
opt = Adam(learning_rate=0.01)
```

```python
modelo.compile(optimizer=opt, loss="binary_crossentropy", metrics=['Accuracy'])
```

```python
loss = modelo.fit(x_normalized, y_output, epochs=100, batch_size=16)
```

```python
for key in loss.history:
  print(key, " - ", loss.history[key])
```
