```python
from matplotlib import pyplot as plt
```

```python
import random
```

```python
areas_imoveis = [random.randint(50, 300) for _ in range(100)]

# Lista com valores dos imóveis em milhares de reais, correlacionados com as áreas
valores_imoveis = [area * random.uniform(5, 10) for area in areas_imoveis]
```

```python
plt.scatter(areas_imoveis, valores_imoveis)
plt.show()
```

```python
def linha(valores, w = 0, b = 0):
    y_hat = []
    for valor in valores:
        y_hat.append((valor * w) + b)
    return y_hat
```

```python
linha_y_hat = linha(areas_imoveis, 8, 0)
```

```python
plt.scatter(areas_imoveis, valores_imoveis)
plt.plot(areas_imoveis, linha_y_hat, color="red" )
plt.show()
```

```python
def subtrair_lista(lista1, lista2):
    nova_lista = []
    for i in range(len(lista1)):
        nova_lista.append(lista1[i] - lista2[i])
    return nova_lista
        
```

```python
def somar_custos( lista ):
    soma = 0
    for valor in lista:
        soma += (valor ** 2)
    return soma
```

```python
inclinacao = -10
for i in range(20):
    linha_y_hat = linha(areas_imoveis, inclinacao, 0)
    custos = subtrair_lista(valores_imoveis, linha_y_hat) 
    soma = somar_custos(custos)
    custo_total = soma / (2 * len(linha_y_hat))
    print(f"Inclinação: {inclinacao}   Custo: {custo_total}")
    inclinacao = inclinacao + 1 
    
```
