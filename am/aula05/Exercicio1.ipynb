## Vendas de maçãs

```python
exportacoes_macas_2023 = {
    "Janeiro": 3000,
    "Fevereiro": 2800,
    "Março": 3200,
    "Abril": 2900,
    "Maio": 3100,
    "Junho": 2700,
    "Julho": 2800,
    "Agosto": 3000,
    "Setembro": 2900,
    "Outubro": 3100,
    "Novembro": 3200,
    "Dezembro": 3000
}
```

```python
nomes_meses = []
valores = []
for item in exportacoes_macas_2023.items():
    nomes_meses.append(item[0])
    valores.append(item[1])
```

```python
plt.barh(nomes_meses, valores)
plt.show()
```
