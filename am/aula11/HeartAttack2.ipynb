```python
import tensorflow as tf
```

```python
import pandas as pd
```

```python
tf.__version__
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Input, Dense
```

```python
from tensorflow.keras.optimizers import RMSprop
```

```python
modelo = Sequential()
```

```python
modelo.add( Input ( (13,) ))
```

```python
modelo.add( Dense(39) )
```

```python
modelo.add( Dense(6) )
```

```python
modelo.add( Dense(1) )
```

```python
modelo.summary()
```

```python
opt = RMSprop(learning_rate=0.01)
```

```python
modelo.compile(optimizer=opt, loss="binary_crossentropy", metrics=['accuracy'])
```

```python
df = pd.read_csv("./heart.csv")
df.head()
```

```python
input_df = df[['age', 'sex', 'cp', 'trtbps', 'chol','fbs', 'restecg', 'thalachh', 'exng', 'oldpeak', 'slp', 'caa', 'thall']]
input_df
```

```python
output_df = df[['output']]
output_df
```

```python
loss = modelo.fit(input_df, output_df, epochs=100, batch_size=16)
```

```python
for key in loss.history:
  print(key, " - ", loss.history[key])
```
