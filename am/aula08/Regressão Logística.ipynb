```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
numeros = np.arange(-10, 10, 0.0001)
```

```python
numeros
```

```python
def sigmoid( x ):
    return 1 / (1 + np.exp(-x))
```

```python
y = sigmoid( numeros )
```

```python
plt.scatter(numeros, y)
plt.show()
```
