# Exercicio, encontrar o melhor valor de W e B em 10 tentativas

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
desvio = np.random.random(size=100)
```

```python
mm_chuva = np.arange(50, 150)
mm_chuva = mm_chuva * (desvio * 0.10)
```

```python
colheita = np.arange(3, 103)
colheita = colheita * (desvio * 0.10)
```

```python
W = 0.7
```

```python
B = -0.8
```

```python
reta = (mm_chuva * W) + B
```

```python
custo = np.mean( (reta - colheita) ** 2 )
custo
```

```python
plt.scatter(mm_chuva, colheita, label="Dados")
plt.scatter(mm_chuva, reta, label="Regress√£o Linear")
plt.legend()
plt.show()
```
