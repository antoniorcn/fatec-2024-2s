```python
def converte_categorico( valor ):
    if valor > 0.5:
        return "positivo"
    else: 
        return "negativo"
```

```python
diabetes_df["categorica"] = diabetes_df["Outcome"].apply(converte_categorico)
```

```python
diabetes_df
```

```python
def faixa_etaria( idade ):
    if idade <= 25:
        return "jovem"
    elif 25 < idade <= 55:
        return "adulto"
    elif 55 < idade <= 80:
        return "senior"
    elif idade > 80:
        return "idoso"
```

```python
diabetes_df["faixa_etaria"] = diabetes_df["Age"].apply(faixa_etaria)
```
