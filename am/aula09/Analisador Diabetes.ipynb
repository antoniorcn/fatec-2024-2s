```python
import pandas as pd
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
diabetes_df = pd.read_csv("../../../dados/diabetes.csv")
```

```python
diabetes_df
```

```python
plt.scatter(diabetes_df["Glucose"], diabetes_df["SkinThickness"], c=diabetes_df["Outcome"])
plt.show()
```

```python
x = diabetes_df[["Glucose", "SkinThickness"]]
```

```python
x_train = x.iloc[0:500]
x_test = x.iloc[500:]
```

```python
y = diabetes_df["Outcome"]
```

```python
y_train = y.iloc[0:500]
y_test = y.iloc[500:]
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict([[140.0, 30.0]])
```
