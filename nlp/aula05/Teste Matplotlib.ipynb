```python
from matplotlib import pyplot as plt
```

```python
# Lista de meses
meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]

# Lista de temperaturas médias correspondentes
temperaturas = [26.5, 26.0, 25.0, 23.0, 20.0, 18.0, 17.0, 19.0, 22.0, 24.0, 25.5, 27.0]
```

```python
plt.pie(temperaturas, labels=meses)
```

```python
plt.plot(temperaturas)
plt.text(6, 18, 'Inverno')
plt.title("Temperaturas médias em 2023 no estado de SP")
plt.ylabel("Temperaturas")
plt.xlabel("Meses")
plt.savefig("temperaturas.png")
plt.show()
```

```python
eixo_y = []
eixo_x = []
for i in range(-200, 200):
    eixo_x.append(i)
    eixo_y.append(i ** 2)
```

```python
plt.plot(eixo_x, eixo_y)
plt.axhline(20000)
plt.axvline(0)
```

```python
y_serie = [2, 4, 1]
x_serie = [1, 2, 3]
y2_serie = [4, 1, 3]
plt.scatter(x_serie, y_serie, color="green", linestyle="dashed", marker = "o", label="Serie Y1")
plt.scatter(x_serie, y2_serie, color="red", linestyle="dashed", marker = "o", label="Serie Y2")
plt.title("My First Graphic")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim(1, 3)
plt.ylim(1, 20)
plt.legend()
```
