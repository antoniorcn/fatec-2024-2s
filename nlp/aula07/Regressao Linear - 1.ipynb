```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
import random
```

```python
x = np.arange(0, 45, 0.5)
```

```python
y = np.array([(random.randint(7, 10) * temperatura) for temperatura in x])
```

```python
a = 2
b = 3
linha_y = (x * a) + b
```

```python
plt.scatter(x, y)
plt.plot(x, linha_y, color="red")
plt.ylim(0, 500)
plt.show()
```
