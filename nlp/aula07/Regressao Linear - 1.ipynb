```python
def funcao_linear( x, w, b ):
    return x * w + b
```

```python
x
```

```python
def funcao_custo( y, y_hat ):
    custos = (y_hat - y) ** 2
    soma = np.sum(custos)
    m = len(y)
    custo_total = soma / (2 * m)
    return custo_total
```

```python
y_hat = funcao_linear( x, 7.9, 0 )
custo = funcao_custo(y_hat, y)  # Pedro
print("Custo do Pedro: ", custo)
plt.scatter(x, y)
plt.plot(x, y_hat, color="red")
plt.ylim(0, 500)
plt.show()
```

```python
y_hat = funcao_linear( x, 9, -3 )
custo = funcao_custo( y_hat, y )
print("Custo Gabriel: ", custo)
plt.scatter(x, y)
plt.plot(x, y_hat, color="red")
plt.ylim(0, 500)
plt.show()
```

```python
def recalcular_w(x, y, w, b, tx_aprendiz):
    m = len(x)
    y_hat = funcao_linear(x, w, b)
    custo_dw = (y_hat - y) * x
    dw = np.sum(custo_dw) / m
    return w - (tx_aprendiz * dw)
```

```python
def recalcular_b(x, y, w, b, tx_aprendiz):
    m = len(x)
    y_hat = funcao_linear(x, w, b)
    custo_db = y_hat - y
    db = np.sum(custo_db) / m
    return b - (tx_aprendiz * db)
```

```python
b = 100
w = 1000
tx_aprendizagem = 0.0001
```

```python
for _ in range(500):
    y_hat = funcao_linear(x, w, b)
    custo = funcao_custo(y_hat, y)
    print(f"W: {w}\tB: {b}\tCusto: {custo}")
    w = recalcular_w(x, y, w, b, tx_aprendizagem)
    b = recalcular_b(x, y, w, b, tx_aprendizagem)
```
